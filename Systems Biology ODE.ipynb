In this notebook we implement a simple ultradian model of the glucose-insulin interactions, as originally given in https://arxiv.org/pdf/2202.01723.pdf . We then use Runge Krutta 4th order to solve it.

The systems of ODEs is given as:

$$
\begin{align}
\frac{dI_p}{dt} &= f_1(G) - E \left( \frac{I_p}{V_p}- \frac{I_i}{V_i} \right) - \frac{I_p}{t_p} \quad& \frac{d h_1}{dt} &= \frac{1}{t_d}\left( I_p - h_1 \right) ,\\
\frac{dI_i}{dt} &= E \left( \frac{I_p}{V_p} - \frac{I_i}{V_i} \right) - \frac{I_i}{t_i}, \quad& \frac{dh_2}{dt} &= \frac{1}{t_d}\left( h_1 - h_2 \right),\\
\frac{dG}{d t} &= f_4(h_3) + I_G(t) - f_2(G) - f_3(I_i)G, \quad& \frac{dh_3}{d t} &= \frac{1}{t_d}\left( h_2 - h_3 \right)
\end{align}
$$
where $I_p$ is the plasma insulin concentration, $I_i$ is the interstitial insulin concentration, $G$ is glucose concentration, and $(h_1,h_2,h_3)$ is a three stage filter to mimic the response of plasma insulin to glucose levels.

Other parameters and functions include:

- $E$ = rate constant for exchange of insulin between plasma and remote compartments
- $I_G$ = exogenous (externally driven) glucose delivery rate
- $t_p$ = time constant for plasma insulin degradation
- $t_i$ = time constant for remote insulin degradation
- $t_d$ = delay time between plasma insulin and glucose production
- $V_p$ = volumne of insulin distribution in the plasma
- $V_i$ = volume of remote insulin compartment
- $V_g$ = volume of glucose space
- $f_1(G)$ = rate of insulin production
- $f_2(G)$ = insulin-independent glucose utilization
- $f_3(I_i)$ = insulin-dependent glucose utilization
- $f_4(h_3)$ = delayed insulin-dependent glucose utilization
- $k$ = decay constant

$f_1, \ldots, f_4, I_g(t), \kappa$ are given by:
$$
\begin{align}
	f_1(G) &= \frac{R_m}{1+ \textrm{exp}\left( -\frac{G}{V_g C_1} + a_1 \right) }, \quad& f_4(h_3) &= \frac{R_g}{1 + \textrm{exp}\left(\alpha \left( \frac{h_3}{C_5V_p} - 1 \right)  \right) },\\
	f_2(G) &= U_b \left( 1- \textrm{exp}\left( -\frac{G}{C_2 V_g} \right)  \right), \quad& I_G(t) &= \sum_{j=1}^{N} m_j k \textrm{exp}\left( k(t_j-t) \right) ,\\
	f_3(I_i) &= \frac{1}{C_3 V_g}\left( U_0 + \frac{U_m}{1+(\kappa I_i)^{-\beta }} \right), \quad& \kappa &= \frac{1}{C_4}\left( \frac{1}{V_i}+ \frac{1}{Et_i} \right)
\end{align}
$$

Our initial condition is $x(0) = \left[12, 4, 110, 0, 0, 0 \right]$ with three nutrition events $(t_j,m_j) = \left[(300,60), (650,40), (1100,50)\right]$.

```python
import numpy as np


def set_parameters(tt = np.arange(0, 1800, 1, dtype=float),
meal_tt = np.array([300, 650, 1100]),
meal_qq = np.array([60e3, 40e3, 50e3]),
Vpp=3,
Vii=11,
Vgg=10,
Ee=0.2,
tpp=6,
tii=100,
tdd=12,
kk=1 / 120,
Rmm=209,
a11=6.6,
C11=300,
C22=144,
C33=100,
C44=80,
C55=26,
Ubb=72,
U00=4,
Umm=90,
Rgg=180,
alphaa=7.5,
betaa=1.772,
y00 = np.array([12.0 * Vp, 4.0 * Vi, 110.0 * Vg, 0.0, 0.0, 0.0]) # Initial condition
):
    
    global t
    global meal_t
    global meal_q
    global Vp
    global Vi
    global Vg
    global E
    global tp
    global ti
    global td
    global k
    global Rm
    global a1
    global C1
    global C2
    global C3
    global C4
    global C5
    global Ub
    global U0
    global Um
    global Rg
    global alpha
    global beta

    global y0
    
    t = tt
    meal_t = meal_tt
    meal_q = meal_qq
    Vp=Vpp
    Vi=Vii
    Vg=Vgg
    E=Ee
    tp=tpp
    ti=tii
    td=tdd
    k=kk
    Rm=Rmm
    a1=a11
    C1=C11
    C2 = C22
    C3=C33
    C4=C44
    C5=C55
    Ub=Ubb
    U0=U00
    Um=Umm
    Rg=Rgg
    alpha=alphaa
    beta=betaa
    
    y0 = y00 # Initial condition

def sbode(t,y): # For a given state and time, we obtain the values for our partial derivatives
    f1 = Rm / (1 + np.exp(-y[2] / Vg / C1 + a1))
    f2 = Ub * (1 - np.exp(-y[2] / Vg / C2))
    kappa = (1 / Vi + 1 / E / ti) / C4
    f3 = (U0 + Um / (1 + (kappa * y[1]) ** (-beta))) / Vg / C3
    f4 = Rg / (1 + np.exp(alpha * (y[5] / Vp / C5 - 1)))
    IG = np.sum(
        meal_q * k * np.exp(k * (meal_t - t)) * np.heaviside(t - meal_t, 0.5)
    )
    tmp = E * (y[0] / Vp - y[1] / Vi)
    return np.array([
        f1 - tmp - y[0] / tp,
        tmp - y[1] / ti,
        f4 + IG - f2 - f3 * y[2],
        (y[0] - y[3]) / td,
        (y[3] - y[4]) / td,
        (y[4] - y[5]) / td,
    ], dtype = float)
```

```python
# Now we apply Runge Krutta 4th order

step = 1
y = np.array([y0])

for i in t[:-1]:
    k1 = step * sbode(i, y[-1])
    k2 = step * sbode(i + step/2, y[-1] + k1/2)
    k3 = step * sbode(i + step/2, y[-1] + k2/2)
    k4 = step * sbode(i + step, y[-1] + k3)
    y = np.concatenate((y, [y[-1] + k1/6 + k2/3 + k3/3 + k4/6]), axis = 0)
    
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3,1, figsize = (16,16))

fig.suptitle('Runge Krutta 4th Order (by me)', y = 0.92)

axs[0].plot(t,y[:,0])
axs[0].set_title('Plasma Insulin Concentration')
axs[0].set_xlabel('time $t$ (mins)')
axs[0].set_ylabel('$I_p$ ( μU/ml )')


axs[1].plot(t,y[:,1])
axs[1].set_title('Interstitial Insulin Concentration')
axs[1].set_xlabel('time $t$ (mins)')
axs[1].set_ylabel('$I_t$ ( μU/ml )')

axs[2].plot(t,y[:,2])
axs[2].set_title('Glucose Concentration')
axs[2].set_xlabel('time $t$ (mins)')
axs[2].set_ylabel('$G$ ( mg/dl )')


```

```python
# Now let's compare to SciPy's ODE solver
from scipy.integrate import odeint

# SciPy's odeint uses lsoda from FORTRAN, which uses either Adams method for nonstiff ODEs, or BDF for stiff ODEs.
# I would guess that in this system it is using the Adams method.

y1 = odeint(sbode,y0,t, tfirst=True)

fig, axs = plt.subplots(3,1, figsize = (16,16))
fig.suptitle('SciPy ODE Solver', y = 0.92)


axs[0].plot(t,y1[:,0])
axs[0].set_title('Plasma Insulin Concentration')
axs[0].set_xlabel('time $t$ (mins)')
axs[0].set_ylabel('$I_p$ ( μU/ml )')


axs[1].plot(t,y1[:,1])
axs[1].set_title('Interstitial Insulin Concentration')
axs[1].set_xlabel('time $t$ (mins)')
axs[1].set_ylabel('$I_t$ ( μU/ml )')

axs[2].plot(t,y1[:,2])
axs[2].set_title('Glucose Concentration')
axs[2].set_xlabel('time $t$ (mins)')
axs[2].set_ylabel('$G$ ( mg/dl )')


```

```python
# And now we plot error

error = np.abs(y-y1)

fig, axs = plt.subplots(3,1, figsize = (16,16))

fig.suptitle('Relative Error between RK4 and SciPy', y = 0.92)

axs[0].plot(t,error[:,0])
axs[0].set_title('Plasma Insulin Concentration Error (to SciPy)')
axs[0].set_xlabel('time $t$ (mins)')
axs[0].set_ylabel('$I_p$ ( μU/ml )')


axs[1].plot(t,error[:,1])
axs[1].set_title('Interstitial Insulin Concentration Error (to SciPy)')
axs[1].set_xlabel('time $t$ (mins)')
axs[1].set_ylabel('$I_t$ ( μU/ml )')

axs[2].plot(t,error[:,2])
axs[2].set_title('Glucose Concentration Error (to SciPy)')
axs[2].set_xlabel('time $t$ (mins)')
axs[2].set_ylabel('$G$ ( mg/dl )')
```

It is clear that the ODE system originally posed is flawed, due to the large increase in glucose levels. More investigation should be done to determine the incorrect factor in the original system.

Nevertheless, both RK4 and SciPy agree to a high degree of accuracy as to the solution to the system of ODEs. This indicates that the numerical solution is likely quite accurate to the true solution of the system.

Now we attempt to work backwards. Given input data, can we determine values for parameters so that our system of ODEs fits the data?

Because the system of ODEs is inherently flawed, we will generate data using the ODE with random parameters, and attempt to work backwards from that data to determine the parameters that were used. In this case, we will allow $C_1 \in (60.0,540)$, $C_2 \in (28.8,259)$, $C_3 \in (20.0,180.0)$,$C_4 \in (16.0,144)$, and $C_5 \in (5.20,46.8)$. We will use the nomimal values (provided in the original initialization) as the initial guess, and optimize from there.

```python
C1 = np.random.normal(300., 80., 1) # We hav a 99.7% chance to be within the interval we would like
C2 = np.random.normal(144., 38.3, 1)
C3 = np.random.normal(100., 26.6, 1)
C4 = np.random.normal(80., 21.3, 1)
C5 = np.random.normal(26., 6.7, 1)

set_parameters(C11 = C1, C22 = C2, C33 = C3, C44 = C4, C55 = C5)

print('Our true parameters are:')
print('C1 = ' + str(C1) + '\n'
     + 'C2 = ' + str(C2) + '\n'
     + 'C3 = ' + str(C3) + '\n'
     + 'C4 = ' + str(C4) + '\n'
     + 'C5 = ' + str(C5) + '\n')

bigtimestep = np.arange(0, 1800, 5, dtype=float)
smalltimestep = np.arange(0, 1800, 1, dtype=float)


y2 = odeint(sbode, y0, t = bigtimestep, tfirst=True) # This will be our input data which we would like to fit to.
# We will assume we are given less timesteps than the original problem, try to fit, and see how well we did when timesteps are increased
y2full = odeint(sbode, y0, t = smalltimestep, tfirst=True)
```

```python
def objFun(guessparam):
    C1, C2, C3, C4, C5 = guessparam
    set_parameters(C11 = C1, C22 = C2, C33 = C3, C44 = C4, C55 = C5)
    guessy = odeint(sbode, y0, t = np.arange(0, 1800, 5, dtype=float), tfirst=True)
    errVal = np.sum(np.square(y2-guessy))
    return errVal



bounds = [(60.,540.), (28.8,259.), (20.,180.), (16.0, 144.0), (5.2, 46.8)]
init_params = [300.,144.,100.,86.,26.]

from scipy.optimize import fmin

bestparams = fmin(objFun, x0 = [300.,144.,100.,86.,26.])
```

```python
C1, C2, C3, C4, C5 = bestparams

set_parameters(C11 = C1, C22 = C2, C33 = C3, C44 = C4, C55 = C5)

bestguessy = odeint(sbode, y0, t = smalltimestep, tfirst=True)

print('Our estimated parameters are:')
print('C1 = ' + str(C1) + '\n'
     + 'C2 = ' + str(C2) + '\n'
     + 'C3 = ' + str(C3) + '\n'
     + 'C4 = ' + str(C4) + '\n'
     + 'C5 = ' + str(C5) + '\n')

print('Total Residual error:')
print(np.sum(np.square(y2full-bestguessy)))
```

```python
fig, axs = plt.subplots(3,1, figsize = (16,16))

fig.suptitle('Estimating Parameters', y = 0.92)

axs[0].plot(smalltimestep,np.abs(y2full[:,0] - bestguessy[:,0]))
axs[0].set_title('Absolute Error in Plasma Insulin Concentration')
axs[0].set_xlabel('time $t$ (mins)')
axs[0].set_ylabel('$I_p$ ( μU/ml )')


axs[1].plot(smalltimestep,np.abs(y2full[:,1] - bestguessy[:,1]))
axs[1].set_title('Absolute Error in Interstitial Insulin Concentration')
axs[1].set_xlabel('time $t$ (mins)')
axs[1].set_ylabel('$I_t$ ( μU/ml )')

axs[2].plot(smalltimestep,np.abs(y2full[:,2] - bestguessy[:,2]))
axs[2].set_title('Absolute Error in Glucose Concentration')
axs[2].set_xlabel('time $t$ (mins)')
axs[2].set_ylabel('$G$ ( mg/dl )')
```

```python
C1, C2, C3, C4, C5 = noisybestparams

set_parameters(C11 = C1, C22 = C2, C33 = C3, C44 = C4, C55 = C5)

bestguessy = odeint(sbode, y0, t = smalltimestep, tfirst=True)

print('Our estimated parameters are:')
print('C1 = ' + str(C1) + '\n'
     + 'C2 = ' + str(C2) + '\n'
     + 'C3 = ' + str(C3) + '\n'
     + 'C4 = ' + str(C4) + '\n'
     + 'C5 = ' + str(C5) + '\n')

print('Total Residual error:')
print(np.sum(np.square(y2full-bestguessy)))
```
